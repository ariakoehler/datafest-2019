```python
def convert_tr(tr_in):
    if type(tr_in) == str:
        return float(tr_in[:-1])/100
    else:
        return tr_in
```

```python
welldf['TrainingReadiness'] = welldf['TrainingReadiness'].apply(convert_tr)
```
