```python
def time_interval(init_time, check_time, interval):
    return True if ((init_time - check_time) <= interval and (init_time - check_time) > timedelta(0)) else False
```

```python
inter = timedelta(days=7)

# print(inter)
# for player in rpe_group.groups:
#     pass

# for day in rpe_group[2]['Date']:
#     print(day)

df_2 = rpe_date.where(rpe_date['PlayerID']==2)

df_2 = df_2.dropna(how='all')

weekdf = pd.DataFrame(columns=['WeekDailyLoad', 'WeekAcuteLoad',
                              'WeekChronicLoad', 'WeekAcuteChronicRatio',
                              'WeekDuration'],
                     index=rpe_date.index)

# for date in df_2['Date']:
#     ind2 = []
#     for date_check in df_2['Date']:
#         if time_interval(date, date_check, inter):
#             ind2.append(date_check)
#     weekdf.append(df_2.fil)

# rpe_with_past = rpe_date

def rpe_past(init_ix):
    window = []
    player = rpe_date['PlayerID'][init_ix]
    init_date = rpe_date['Date'][init_ix]
    for ind in rpe_date[rpe_date['PlayerID'] == player].index:# and \
            # time_interval(init_date, rpe_date[ind]['Date'], inter)]:
        if time_interval(init_date, rpe_date['Date'][ind], inter):
            window.append(ind)
    means = rpe_date.filter(['DailyLoad','AcuteLoad','ChronicLoad',
                     'AcuteChronicRatio','Duration']).iloc[window].mean()
#     means = cols.filter(window, axis=1).mean()
    print(means)
    weekdf.iloc[init_ix] = means
```
